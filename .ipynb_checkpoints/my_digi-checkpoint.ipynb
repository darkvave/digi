```python
fig = plt.figure(figsize=(20,5))
plt.bar(x=data['Name of State / UT'],height=data['Active'],color='blue',width=0.2)
plt.xlabel('STATES')
plt.ylabel('ACTIVE CASES IN EACH STATES')
plt.title('Coronavirus Cases')
plt.xticks(rotation=80)
plt.show()





fig = plt.figure(figsize=(20,5))
plt.bar(x=data['Name of State / UT'],height=data['Deaths'],color='maroon',width=0.2)
plt.xlabel('STATES')
plt.ylabel('DEATHS IN EACH STATES')
plt.title('Coronavirus Cases')
plt.xticks(rotation=80)
plt.show()





fig = plt.figure(figsize=(20,5))
plt.bar(x=data['Name of State / UT'],height=data['Cured/Discharged/Migrated'],color='green',width=0.2)
plt.xlabel('STATES')
plt.ylabel('CURED / DISCHARGED / MIGRATED CASES IN EACH STATES')
plt.title('Coronavirus Cases')
plt.xticks(rotation=80)
plt.show()
```

```python
top_state= data.sort_values('Deaths',ascending=False)
```

```python
top=top_state.head()
```

```python
fig = plt.figure(figsize=(10,5))
plt.bar(x=top['Name of State / UT'],height=top['Deaths'],color='maroon',width=0.2)
#plt.xticks(rotation=80)
plt.title('Top States of Death in Covid')
plt.show()
```

```python
plt.figure(figsize=(20,5))
#plt.subplot(221)
sns.relplot(x='Name of State / UT', y='Deaths',data= top,aspect=2,hue='Deaths',size='Deaths',legend='full')
plt.show()
```

```python
total = 'Cured/Discharged', 'Confirmed Cases', 'Deaths', 'Active'
number = [total_cured, total_confirmed_cases, total_death, total_active]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)
plt.figure(figsize=(20,5))
plt.pie(number, explode=explode, labels=total, colors=colors,autopct='%1.1f%%', shadow=True,startangle=140,rotatelabels=True,textprops={'color':'blue'})
plt.axis('equal')
plt.show()
```
